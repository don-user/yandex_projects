### Проверим статистические показатели и взаимосвязи данных

Проверим как коррелируются между собой признаки, а так же общую картину распределения признаков в разрезе категорий ушел/не ушел из банка:

```python
sns.pairplot(df, hue='exited', markers=['o', 's'])
plt.suptitle(dict_name.get('distribution_main_features'), y=1.02, fontsize=20, fontweight='bold') 
plt.show()
```

### Обучаем модель без учета дисбаланса в выборке

#### Обучаем модель логистической регрессии на стандартных гипперпараметрах

Обучение на дисбалансированной выборке будем производить без настройки гиперпараметров, если сбалансированная выборка даст лучшие результаты, то на сбалансированной выборке настроим гиперпараметры.

```python
#функция для отображения хитмэп
def heatmap_helper(target, predict):
    plt.figure(figsize = (7,6))
    sns.heatmap(confusion_matrix(target, predict), annot=True, fmt='d')
    plt.title(dict_name.get('heatmap_matrix'), fontweight="bold")
    plt.xlabel(dict_name.get('predicted'))
    plt.ylabel(dict_name.get('target'))
    plt.show()
```

```python
#функция для обучения модели, предсказания и отображения основных метрик
def model_train(regresor, 
                features_t = features_train, 
                target_t = target_train, 
                features_v = features_valid, 
                target_v = target_valid,
                heatmap_ind = 0,
                return_ind = True):
    model = regresor
    model.fit(features_t, target_t)
    predicted = model.predict(features_v)
    
    proba_valid = model.predict_proba(features_v)
    
    if heatmap_ind == 0:
        heatmap_helper(target_v, predicted)
        
    print(f'Accuracy {model.__class__.__name__}: ', model.score(features_v, target_v))
    print(f'Recall {model.__class__.__name__}: ', round(recall_score(target_v, predicted), 3))
    print(f'Precision {model.__class__.__name__}: ', round(precision_score(target_v, predicted), 3))
    print(f'F1 {model.__class__.__name__}: ', round(f1_score(target_v, predicted), 3))
    print(f'AUC_ROC {model.__class__.__name__}:', round(roc_auc_score(target_v, proba_valid[:, 1]), 3))
    
    if return_ind:
        return (model.__class__.__name__,
                model.score(features_v, target_v),     
                round(recall_score(target_v, predicted), 3),
                round(precision_score(target_v, predicted), 3),
                round(f1_score(target_v, predicted), 3),
                round(roc_auc_score(target_v, proba_valid[:, 1]), 3))
```

```python
#функция возрашает датафрейм
def df_helper(func, regressor_name):
    class_name, accuracy, recall, precision, f1, auc_roc = func
    return pd.DataFrame(
        [[class_name, regressor_name, accuracy, recall, precision, f1, auc_roc]], 
        columns=['class_name', 'method', 'accuracy', 'recall', 'precision', 'f1', 'auc_roc'])   
```

```python
#храним обощенные данные по метрикам моделей
df_res = pd.DataFrame(columns=['class_name', 'method', 'accuracy', 'recall', 'precision', 'f1', 'auc_roc'])
```

### Проверим результаты при баланисровании выборки методом upsampling

```python
#функция баланисровки выборки upsample и downsample
def change_sampled(type_sample, repeat=3, fraction=0.2, features=features_train, target=target_train):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    
    if type_sample == 'upsample':
        features_sampled = pd.concat([features_zeros] + [features_ones] * repeat)
        target_sampled = pd.concat([target_zeros] + [target_ones] * repeat)
    elif type_sample == 'downsample':
        features_sampled = (pd.concat([features_zeros.
                                           sample(frac=fraction, random_state=12345)] + [features_ones]))
        target_sampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    else:
        print('Ошибка выбора функции!')
        
    features_sampled, target_sampled = shuffle(
        features_sampled, target_sampled, random_state=12345)
    
    return features_sampled, target_sampled
```

Получим выборку сбалансированную с помощью `upsample`

```python
features_upsampled, target_upsampled = change_sampled('upsample')
print('features_upsampled -', features_upsampled.shape, 'target_upsampled -', target_upsampled.shape)
```

### Подборка гипперпараметров для алгоритмов

Подбор гипперпараметров будем производить с помощью `GridSearchCV`, при определении лучшей модели оценивать будет при помощи 2-х метрик, `F1 и AUC`. Оценку `AUC` будем проводить по шкале: 
- 0.8 <= auc <= 1 — *модель работает превосходно*
- 0.6 <= auc <= 0.8 — *модель работает хорошо*
- 0.5 <= auc <= 0.6 — *модель работает удовлетворительно*
- auc <= 0.5 — *модель не работает*.

#### Подбор гипперпараметров для алгоритма LogisticRegression

```python
#функция отображения графика auc_roc
def auc_roc_helper(best_estimator, features=features_valid, target=target_valid):
    
    proba_valid = best_estimator.predict_proba(features)
    fpr, tpr, thresholds = roc_curve(target, proba_valid[:, 1])
    
    plt.figure(figsize=(6, 6))
    plt.plot(fpr, tpr)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.0])
    plt.xlabel(dict_name.get('FPR'))
    plt.ylabel(dict_name.get('TPR'))
    plt.title(f'{best_estimator.__class__.__name__}: {dict_name.get("ROC")}', fontweight='bold')
    plt.plot([0, 1], [0, 1], linestyle='--')
    plt.show()
```

```python
#функция по работе с GridSearchCV и отображению основных показателей
def grid_helper(grid, features=features_train, target=target_train,):
    
    grid.fit(features, target)
    
    print('Параметры лучшей модели:', grid.best_estimator_) 
    print('Лучший f1 на обучающей выборке:', grid.best_score_.round(3))
```

```python
#храним обощенные данные по метрикам моделей 
df_res_auc = pd.DataFrame(columns=['class_name', 'method', 'accuracy', 'recall', 'precision', 'f1', 'auc_roc'])
```

```python
param_lr = [{'penalty':['l1', 'l2']}, 
            {'C':[0.001, 0.01, 0.1, 1, 10, 100]},
            {'intercept_scaling': [0.5, 1.0, 1.5]}]
clf = LogisticRegression(
    random_state=12345, 
    solver='liblinear',
    class_weight='balanced')
                   
grid_lr = GridSearchCV(estimator = clf,
                       param_grid = param_lr, 
                       scoring = 'f1', 
                       cv = 5, 
                       verbose=0)
grid_helper(grid_lr)
```
