### Проверка корректности расчета эффективности обогащения чернового концентрата

```python
#расчет эффетивности обогащения по формуле
recovery_calc = (
    (df_train['rougher.output.concentrate_au'] * 
     (df_train['rougher.input.feed_au'] - df_train['rougher.output.tail_au'])) / 
    (df_train['rougher.input.feed_au'] * 
     (df_train['rougher.output.concentrate_au'] - df_train['rougher.output.tail_au'])) * 100
)
print(recovery_calc.sum() == df_train['rougher.output.recovery'].sum())
```

```python
print(mean_absolute_error(recovery_calc, df_train['rougher.output.recovery']))
```

На тренировочной выборке результаты расчетов эффективности обогащения для признака `rougher.output.recovery`, расчитаны корректно, суммы результатов расчитаного по формуле показателя и в таблице равны, средняя абсолютная ошибка составляет очень низкое значение, и практически стремиться к нулю.

### Анализ признаков отсутвующих в тестовой выборке

```python
# функция выявление признаков с сильной корреляцией
def higth_corr(df):
    cr = df.corr().abs().unstack().sort_values(ascending=False)
    
    cr = cr[(cr >= 0.7) & (cr < 1)]
    return(cr
           .to_frame()
           .reset_index()
           .rename(columns={'level_0' : 'feature_1', 'level_1' : 'feature_2', 0 : 'corr'})
           [::2])
```

### Концентрация металлов (Au, Ag, Pb) на различных этапах очистки

Создадим таблицу в которой будут содержаться информация о средней концентрации метталов `'pb'`, `'ag'`, `'au'` на разных стадиях обработки.

```python
met_concentrate = df_full[
    [i for i in df_full.columns if i.split('.')[-1] in ['concentrate_ag', 'concentrate_pb', 'concentrate_au',
                                                        'feed_ag', 'feed_pb','feed_au']]
]
```

```python
metals = ['pb', 'ag', 'au']
state_list = [[x for x in met_concentrate.columns if x.split('_')[-1] == v] for v in metals]
ax = [311, 312, 313]
plt.figure(figsize = (14, 14))
plt.suptitle('Гистограммы металлов по стадиям обработки', fontsize = 14, fontweight='bold', y = 0.95)

for i, v in enumerate(state_list):
    plt.subplot(ax[i])
    sns.histplot(data=met_concentrate[v], legend=True)
 
    plt.title(f'Концентрация {metals[i]}', fontweight="bold")
    plt.legend(bbox_to_anchor=(1,1), loc='upper left', labels=v)

plt.show()
```

```python
met_concentrate = (met_concentrate.agg('mean', axis=0)
                  .to_frame().reset_index()
                  .rename(columns={'index' : 'state', 0 : 'concentrate'}))
met_concentrate['metals'] = met_concentrate['state'].apply(lambda x: x[-2:])
met_concentrate['state'] = met_concentrate['state'].apply(
    lambda x: 'start' if x.split('.')[1] == 'input' else x.split('.')[0])
met_concentrate = met_concentrate.sort_values(by='state', ascending=False, ignore_index=True)
display(met_concentrate)
```

Для анализа визуализируем таблицу в разрезе металлов и по этапам обработки

```python
ax = [131, 132, 133]
color_list = ['Set2', 'rocket', 'YlOrBr']
plt.figure(figsize = (18,5))
plt.suptitle('Концентрация металлов по стадиям обработки', fontsize = 14, fontweight='bold', y = 1.05)
for i, v in enumerate(['pb', 'ag', 'au']):
    plt.subplot(ax[i])
    sns.lineplot(data=met_concentrate.query('metals == @v'), markers=True, palette=color_list[i])
    
    plt.xticks(
        met_concentrate.query('metals == @v').index, 
        labels=met_concentrate.query('metals == @v')['state'])
    plt.title(f'Концентрация {v}', fontweight="bold")
```

### Исследования сумарных концентраций веществ в сырье в черновом и финальном концентратах

```python
#функция по отображению графиков hist и violinplot
def concentrate_sum(state_name, color_1, color_2):
    sum_col = [col for col in df_full.columns if ((col.split('_')[-1] in ['au', 'ag', 'pb', 'sol']) &
                                                 (col.startswith(state_name)))]
    
    sum_col = df_full[sum_col].sum(axis=1)
    
    plt.figure(figsize = (20,7))
    plt.suptitle(f'Стадия концентрации: {state_name}', fontsize = 14, fontweight='bold', y = 1)
    plt.subplot(121)
    sns.histplot(sum_col, bins=70, color=color_1)
    
    plt.subplot(122)
    sns.violinplot(data=sum_col, color=color_2, bw=.2, cut=1, linewidth=1)
    plt.show()
    
    return sum_col
```

```python
df_sum_col = pd.DataFrame({'date' : df_full['date']})
```

```python

df_sum_col['before_float'] = concentrate_sum('rougher.input.feed', 'm', 'fuchsia')
```

```python
df_sum_col['after_float'] = concentrate_sum('rougher.output.concentrate', 'c', 'skyblue')
```

```python
df_sum_col['after_cleaner'] = concentrate_sum('primary_cleaner.output.concentrate', 'darkorange', 'orange')
```

## Модель

### Функция для вычисления итоговой sMAPE.

Подготовим функции для определения качества модели в соотвествии с заданием.

```python
#Функции по оценки качества sMAPE
def s_mape(target, pred):
    return np.mean(abs(target - pred) / ((abs(target) + abs(pred)) / 2)) * 100     
```

```python
#функция для итоговой оценки качества 
def s_mape_final(rougher, final):
    return 0.25 * rougher + 0.75 * final
```

### Обучение моделей

Раздели выборку на основную и целевую, для модели `rougher` выделим отдельные признаки, т.к. в данных `final` могут *содержаться данные, которые не должны быть доступны на этапе* **флотации**. А для стадии `final` данные `rougher` оставим т.к. они известны на этапе процедур `final`.

```python
#функция для кроссвалидации и расчета sMAPE
def final_s_mape(model_rougher, model_final, cv=False):
    
    if cv:
        pipeline_rougher = make_pipeline(StandardScaler(), model_rougher)  
        rougher = cross_val_score(pipeline_rougher, 
                                  feature_train_rougher, 
                                  target_train_rougher, 
                                  scoring=s_mape_score, 
                                  cv=5).mean()
        print(f'{target_train_rougher.name.split(".")[0]} sMape', rougher)
        
        pipeline_final = make_pipeline(StandardScaler(), model_final)  
        final = cross_val_score(pipeline_final, 
                                feature_train_final, 
                                target_train_final, 
                                scoring=s_mape_score, 
                                cv=5).mean()
        print(f'{target_train_final.name.split(".")[0]} sMape', final)
        print('Итоговая оценка sMAPE:',round(s_mape_final(rougher, final), 3))
    else:
        print('Итоговая оценка sMAPE:',round(s_mape_final(model_rougher, model_final), 3))
```

```python
#функция для получения модели
def find_model(regressor, feature, target, param={}):
    pipeline = make_pipeline(StandardScaler(), regressor)
    pipeline.get_params().keys()
    clf = RandomizedSearchCV(
        estimator=pipeline, 
        param_distributions = param, 
        cv = 5, 
        n_jobs=-1,
        scoring=s_mape_score,
        random_state=STATE)    
    clf.fit(feature, target)
    print(f'{target.name.split(".")[0]} sMape', clf.best_score_)
   
    return clf.best_estimator_, clf.best_score_
```

```python
#функция для предсказывания и выводы финального smape
def pred_and_s_mape_print(rougher, final):
    predictions_rougher = rougher.predict(feature_test_rougher)
    predictions_final = final.predict(feature_test_final)
    print('Итоговая оценка sMAPE на тестовой выборке:', 
          round(s_mape_final(s_mape(target_test_rougher, predictions_rougher), 
                             s_mape(target_test_final, predictions_final)), 3))
```

```python
#отрисовка графика
def plot_helper(data, dict_name, col_name):
    data.plot(kind='bar',
              figsize=(12, 6), 
              grid=True, 
              x=col_name, 
              rot=90)
    plt.title(dict_name, fontweight='bold')
    plt.legend(bbox_to_anchor=(1,1), loc='upper left')
    plt.show()
```
